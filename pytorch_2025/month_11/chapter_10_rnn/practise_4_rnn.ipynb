# RNN

```python
import torch
from d2l import torch as d2l

# 令 x 表示一个长度为 n 的序列（batch size=1），每个元素一个输入特征
# 比如 n=3，输入维度=4
X = torch.normal(0, 1, (1, 3, 4))   # (batch_size, num_steps, input_size)
# 假设隐藏层大小为 h=5
W_xh = torch.normal(0, 1, (4, 5))   # (input_size, hidden_size)
H = torch.normal(0, 1, (1, 5))      # (batch_size, hidden_size)
W_hh = torch.normal(0, 1, (5, 5))   # (hidden_size, hidden_size)

# 取序列第一个时间步
X_t = X[:, 0, :]   # (batch_size, input_size)
# RNN的一步
Y = torch.matmul(X_t, W_xh) + torch.matmul(H, W_hh)
Y
```
