```python
class MyClass:
    count = 0
    
    def __init__(self) -> None:
        pass
    
    @classmethod
    def get_count(cls):
        return cls.count
    
    
    @staticmethod
    def add(x, y):
        return x+y
    
obj1 = MyClass()
obj2 = MyClass()

print(obj1.get_count())
MyClass.count = 10
print(obj1.get_count())
print(MyClass.add(1.2,0))
```
