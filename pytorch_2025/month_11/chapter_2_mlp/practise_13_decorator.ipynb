# 装饰器

```python
def my_decorator(func):
    def wrap():
        print("函数执行前")
        func()
        print("函数执行后")
    
    return wrap

@my_decorator
def hello():
    print("hello")

hello()
```

```python
def my_decorator(func):
    def wrapper(*args, **kargs):
        print(f"调用函数的名称{func.__name__}")
        result = func(*args, **kargs)
        print(f"函数返回{result}")
        return result
    
    return wrapper


@my_decorator
def add(x,y):
    return x+y

add(1,2)
```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} 耗时: {end - start:.4f}秒")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(1)
    return "完成"

slow_function()
```

```python
def repeat(times):
    def decorator(func):
        def wrapper(*args, **kargs):
            for _ in range(times):
                result =func(*args, **kargs)
            print(f"执行了{times}次")
            return result
        return wrapper
    return decorator


@repeat(3)
def hello(name):
    print(f"hello {name}")
    
hello("嘿嘿")
```
