# 填充与步幅

```python
import torch
from torch import nn
```

```python
def comp_conv2d(conv2d, X):
    X = X.reshape((1,1)+X.shape)
    Y=conv2d(X)
    return Y.reshape(Y.shape[2:])

conv2d = nn.Conv2d(1,1,kernel_size=3, padding=1)
X=torch.rand(size=(8,8))
comp_conv2d(conv2d, X).shape
```

```python
conv2d = torch.nn.Conv2d(1,1,kernel_size=(5,3), padding=(2,1))
# 8*8 -5*3 +1*1 +4*2 = 8*8
comp_conv2d(conv2d=conv2d, X=X).shape
```

```python
conv2d = nn.Conv2d(1,1,kernel_size=3, padding=1, stride=2)
comp_conv2d(conv2d, X).shape
```

```python
conv2d = nn.Conv2d(1,1,kernel_size=(3,5), padding=(0,1), stride=(3,4))
# 这里的计算公式不正确，卷积后输出大小应该分别对高和宽单独计算：
# H_out = floor((H_in + 2*padding[0] - kernel_size[0] + stride[0]) / stride[0])
# W_out = floor((W_in + 2*padding[1] - kernel_size[1] + stride[1]) / stride[1])
# 代入本例：H_out = floor((8 + 2*0 - 3 + 3) / 3) = floor(8/3) = 2
#          W_out = floor((8 + 2*1 - 5 + 4) / 4) = floor(9/4) = 2
# 即输出形状为(2,2)
comp_conv2d(conv2d, X).shape
```
